## Acquisition et PrÃ©paration

```python
# Ãtape 1: Installer la bibliothÃ¨que
!pip install ccxt pandas

# Ãtape 2: Importer les bibliothÃ¨ques
import ccxt
import pandas as pd
from datetime import datetime

# --- Ãtape 3: Initialiser l'Ã©change et les paramÃ¨tres ---
exchange = ccxt.binance() # On choisit la plateforme Binance
symbol = 'BTC/USDT'      # Le symbole pour Bitcoin vs Tether (trÃ¨s similaire Ã  USD)
timeframe = '1d'         # Intervalle journalier
since_string = '2018-01-01T00:00:00Z' # Format de date ISO8601

print(f"Connexion Ã  {exchange.id} pour tÃ©lÃ©charger les donnÃ©es de {symbol}...")

# --- Ãtape 4: Conversion de la date en millisecondes (requis par l'API) ---
since_milliseconds = exchange.parse8601(since_string)

# --- Ãtape 5: Boucle pour rÃ©cupÃ©rer toutes les donnÃ©es par morceaux ---
all_ohlcv = [] # Liste pour stocker tous les morceaux de donnÃ©es

while True:
    try:
        # RÃ©cupÃ¨re un morceau de donnÃ©es (par dÃ©faut, 500 derniÃ¨res bougies)
        ohlcv = exchange.fetch_ohlcv(symbol, timeframe, since=since_milliseconds)
        
        # Si aucune donnÃ©e n'est retournÃ©e, on a tout rÃ©cupÃ©rÃ©
        if len(ohlcv) == 0:
            break
            
        all_ohlcv += ohlcv # Ajoute le morceau Ã  notre liste
        
        # On met Ã  jour le point de dÃ©part pour la prochaine requÃªte
        # On prend la date de la derniÃ¨re bougie + 1ms pour Ã©viter les doublons
        since_milliseconds = ohlcv[-1][0] + 1

    except Exception as e:
        print(f"Une erreur est survenue pendant le tÃ©lÃ©chargement : {e}")
        break

print("\nâ TÃ©lÃ©chargement de toutes les donnÃ©es terminÃ©.")

# --- Ãtape 6: CrÃ©ation et nettoyage du DataFrame Pandas ---
if all_ohlcv:
    # On crÃ©e le DataFrame
    df = pd.DataFrame(all_ohlcv, columns=['Timestamp', 'Open', 'High', 'Low', 'Close', 'Volume'])
    
    # On convertit le timestamp en format de date lisible
    df['Date'] = pd.to_datetime(df['Timestamp'], unit='ms')
    
    # On met la colonne 'Date' comme index du DataFrame
    df.set_index('Date', inplace=True)
    
    # On supprime la colonne 'Timestamp' qui est maintenant redondante
    df.drop('Timestamp', axis=1, inplace=True)

    print("\n--- AperÃ§u des 5 premiÃ¨res lignes de donnÃ©es ---")
    print(df.head())

    print("\n--- AperÃ§u des 5 derniÃ¨res lignes de donnÃ©es ---")
    print(df.tail())

    print("\n--- Informations sur le DataFrame ---")
    df.info()
else:
    print("\nAucune donnÃ©e n'a Ã©tÃ© tÃ©lÃ©chargÃ©e.")
```

```python
import numpy as np

#DataFrame 'df est dÃ©jÃ  chargÃ© et contient la colonne 'Close'.

# Ãtape 1 : Calculer les rendements logarithmiques
df['log_return'] = np.log(df['Close'] / df['Close'].shift(1))

# Ãtape 2 : Afficher les premiÃ¨res lignes pour vÃ©rifier le rÃ©sultat
print("--- DataFrame avec la nouvelle colonne 'log_return' ---")
print(df.head())

# Ãtape 3 (Optionnel mais recommandÃ©) : Supprimer la premiÃ¨re ligne qui contient une valeur nulle (NaN)
df.dropna(inplace=True)

print("\n--- DataFrame aprÃ¨s suppression de la valeur nulle ---")
print(df.head())
```

### Visualiser les sÃ©ries de prix et de rendements pour observer les rÃ©gimes de marchÃ©.

```python
import matplotlib.pyplot as plt
import seaborn as sns


# --- Configuration du style des graphiques ---
sns.set_style("darkgrid")
plt.rc('figure', figsize=(14, 8))
plt.rc('font', size=12)

# --- CrÃ©ation de la figure et des sous-graphiques ---
# On crÃ©e une figure avec 2 graphiques l'un au-dessus de l'autre (2 lignes, 1 colonne)
# sharex=True signifie qu'ils partageront le mÃªme axe des x (le temps)
fig, ax = plt.subplots(2, 1, sharex=True)

# --- Graphique 1 : Prix de ClÃ´ture ---
ax[0].plot(df['Close'], label='Prix de ClÃ´ture', color='skyblue')
ax[0].set_title('Historique des Prix de ClÃ´ture du Bitcoin (BTC-USD)')
ax[0].set_ylabel('Prix en USD')
ax[0].legend()

# --- Graphique 2 : Rendements Logarithmiques ---
ax[1].plot(df['log_return'], label='Rendements Logarithmiques', color='salmon')
ax[1].set_title('Rendements Logarithmiques Journaliers du Bitcoin')
ax[1].set_ylabel('Rendement Log')
ax[1].set_xlabel('Date')
ax[1].legend()

# Ajouter une ligne horizontale Ã  zÃ©ro pour mieux voir les rendements positifs/nÃ©gatifs
ax[1].axhline(0, color='grey', linestyle='--')

# --- Affichage ---
plt.tight_layout() # Ajuste automatiquement les graphiques pour qu'ils ne se chevauchent pas
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
import numpy as np

# --- 1. Calculer les paramÃ¨tres pour la courbe normale ---
# La courbe normale thÃ©orique doit avoir la mÃªme moyenne et le mÃªme Ã©cart-type que nos donnÃ©es
mu = df['log_return'].mean()
sigma = df['log_return'].std()

# --- 2. CrÃ©er le graphique ---
plt.figure(figsize=(12, 7))

# Tracer l'histogramme des rendements rÃ©els
sns.histplot(df['log_return'], 
             bins=100, 
             kde=True,  # Ajoute une estimation lissÃ©e de la distribution
             stat="density", 
             color='salmon',
             label='Distribution des Rendements RÃ©els')

# Tracer la courbe de la distribution normale thÃ©orique
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000) # CrÃ©e une plage de valeurs pour la courbe
pdf = norm.pdf(x, mu, sigma) # Calcule la densitÃ© de probabilitÃ© pour chaque valeur
plt.plot(x, pdf, 'k--', linewidth=2, label='Distribution Normale ThÃ©orique')

# --- 3. AmÃ©liorer la lisibilitÃ© du graphique ---
plt.title('Distribution des Rendements Logarithmiques du Bitcoin vs. Distribution Normale')
plt.xlabel('Rendement Logarithmique Journalier')
plt.ylabel('DensitÃ©')
plt.legend()
plt.show()

# --- 4. Calculer le Kurtosis ---
kurtosis_value = df['log_return'].kurtosis()
print(f"Le kurtosis des rendements est : {kurtosis_value:.2f}")
if kurtosis_value > 3:
    print("Un kurtosis > 3 confirme la prÃ©sence de 'queues de distribution Ã©paisses' (leptokurtosis).")
```

### Tester la stationnaritÃ© des rendements (test ADF).

```python
pip install statsmodels
```

```python
from statsmodels.tsa.stattools import adfuller


# --- 1. SÃ©lectionner la sÃ©rie Ã  tester ---
# Le test ADF s'applique sur une sÃ©rie (une seule colonne)
returns_series = df['log_return']

# --- 2. RÃ©aliser le test ADF ---
print("RÃ©alisation du test Augmented Dickey-Fuller (ADF)...")
# L'option autolag='AIC' permet Ã  la fonction de choisir le nombre de retards optimal
adf_result = adfuller(returns_series, autolag='AIC')

# --- 3. Afficher les rÃ©sultats de maniÃ¨re claire ---
print(f'ADF Statistic: {adf_result[0]}')
print(f'p-value: {adf_result[1]}')
print('Critical Values:')
for key, value in adf_result[4].items():
    print(f'\t{key}: {value}')

# --- 4. InterprÃ©ter le rÃ©sultat automatiquement ---
print("\n--- Conclusion du test ---")
p_value = adf_result[1]
if p_value <= 0.05:
    print(f"P-value ({p_value:.4f}) <= 0.05. On rejette l'hypothÃ¨se nulle.")
    print("Conclusion : La sÃ©rie des rendements est trÃ¨s probablement STATIONNAIRE.")
else:
    print(f"P-value ({p_value:.4f}) > 0.05. On ne peut pas rejeter l'hypothÃ¨se nulle.")
    print("Conclusion : La sÃ©rie des rendements est trÃ¨s probablement NON-STATIONNAIRE.")
```

## ModÃ©lisation Comparative :

### Model 1 : 30 jours

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. DÃ©finir les paramÃ¨tres ---
window_size = 30 # La taille de notre fenÃªtre glissante (30 jours)

# --- 2. Calculer la volatilitÃ© historique sur la fenÃªtre glissante ---
# .rolling(window=window_size) crÃ©e la fenÃªtre glissante.
# .std() calcule l'Ã©cart-type (la volatilitÃ©) Ã  l'intÃ©rieur de chaque fenÃªtre.
df['historical_volatility'] = df['log_return'].rolling(window=window_size).std()

# --- 3. Annualiser la volatilitÃ© ---
# On multiplie la volatilitÃ© journaliÃ¨re par la racine carrÃ©e du nombre de jours dans une annÃ©e.
# Pour les cryptos, qui tradent 365 jours par an, on utilise 365.
annualization_factor = np.sqrt(365)
df['historical_volatility_annualized'] = df['historical_volatility'] * annualization_factor

# --- 4. Afficher les rÃ©sultats ---
print("--- AperÃ§u du DataFrame avec la volatilitÃ© historique ---")
# Les premiÃ¨res lignes seront NaN car il faut 30 jours de donnÃ©es pour le premier calcul
print(df.tail())

# --- 5. Visualiser la volatilitÃ© historique ---
plt.figure(figsize=(14, 7))
plt.plot(df['historical_volatility_annualized'], label='VolatilitÃ© Historique Annuelle (30 jours)', color='purple')
plt.title('VolatilitÃ© Historique Glissante du Bitcoin')
plt.xlabel('Date')
plt.ylabel('VolatilitÃ© AnnualisÃ©e')
plt.legend()
plt.show()
```

### ModÃ¨le 2 : Estimer un modÃ¨le GARCH(1,1) sur les rendements.

```python
pip install arch
```

```python
from arch import arch_model
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. PrÃ©parer les donnÃ©es ---
# On multiplie souvent les rendements par 100 pour aider le solveur du modÃ¨le Ã  converger
returns = df['log_return'] * 100

# --- 2. SpÃ©cifier et estimer le modÃ¨le GARCH(1,1) ---
# vol='Garch' : spÃ©cifie un modÃ¨le GARCH
# p=1, q=1 : correspond Ã  l'ordre ARCH(1) et GARCH(1)
model = arch_model(returns, vol='Garch', p=1, q=1)

# Estimer le modÃ¨le sur les donnÃ©es
# update_freq=5 affiche l'Ã©tat de l'optimisation toutes les 5 itÃ©rations
results = model.fit(update_freq=5)

# --- 3. Afficher le rÃ©sumÃ© des rÃ©sultats ---
print("\n--- RÃ©sumÃ© du ModÃ¨le GARCH(1,1) ---")
print(results.summary())

# --- 4. Visualiser la volatilitÃ© prÃ©dite par le modÃ¨le ---
fig = results.plot(annualize='D')
plt.suptitle('VolatilitÃ© Conditionnelle EstimÃ©e par GARCH(1,1)')
plt.show()
```

### modÃ¨le 3 : Estimer un modÃ¨le GJR-GARCH(1,1)

```python
from arch import arch_model
import matplotlib.pyplot as plt

# On rÃ©utilise les rendements multipliÃ©s par 100.
returns = df['log_return'].dropna() * 100

# --- 1. SpÃ©cifier et estimer le modÃ¨le GJR-GARCH(1,1) ---
# Le changement est ici : on ajoute o=1 pour inclure le terme asymÃ©trique (gamma)
# p=1 : Terme GARCH (beta)
# o=1 : Terme AsymÃ©trique (gamma)
# q=1 : Terme ARCH (alpha)
model_gjr = arch_model(returns, vol='Garch', p=1, o=1, q=1)

# Estimer le modÃ¨le
results_gjr = model_gjr.fit(update_freq=5)

# --- 2. Afficher le rÃ©sumÃ© des rÃ©sultats ---
print("\n--- RÃ©sumÃ© du ModÃ¨le GJR-GARCH(1,1) ---")
print(results_gjr.summary())

# --- 3. Visualiser la volatilitÃ© prÃ©dite ---
fig = results_gjr.plot(annualize='D')
plt.suptitle('VolatilitÃ© Conditionnelle EstimÃ©e par GJR-GARCH(1,1)')
plt.show()
```
