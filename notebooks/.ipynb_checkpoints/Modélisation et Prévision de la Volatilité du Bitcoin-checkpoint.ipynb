## Acquisition et Préparation

```python
# Étape 1: Installer la bibliothèque
!pip install ccxt pandas

# Étape 2: Importer les bibliothèques
import ccxt
import pandas as pd
from datetime import datetime

# --- Étape 3: Initialiser l'échange et les paramètres ---
exchange = ccxt.binance() # On choisit la plateforme Binance
symbol = 'BTC/USDT'      # Le symbole pour Bitcoin vs Tether (très similaire à USD)
timeframe = '1d'         # Intervalle journalier
since_string = '2018-01-01T00:00:00Z' # Format de date ISO8601

print(f"Connexion à {exchange.id} pour télécharger les données de {symbol}...")

# --- Étape 4: Conversion de la date en millisecondes (requis par l'API) ---
since_milliseconds = exchange.parse8601(since_string)

# --- Étape 5: Boucle pour récupérer toutes les données par morceaux ---
all_ohlcv = [] # Liste pour stocker tous les morceaux de données

while True:
    try:
        # Récupère un morceau de données (par défaut, 500 dernières bougies)
        ohlcv = exchange.fetch_ohlcv(symbol, timeframe, since=since_milliseconds)
        
        # Si aucune donnée n'est retournée, on a tout récupéré
        if len(ohlcv) == 0:
            break
            
        all_ohlcv += ohlcv # Ajoute le morceau à notre liste
        
        # On met à jour le point de départ pour la prochaine requête
        # On prend la date de la dernière bougie + 1ms pour éviter les doublons
        since_milliseconds = ohlcv[-1][0] + 1

    except Exception as e:
        print(f"Une erreur est survenue pendant le téléchargement : {e}")
        break

print("\n✅ Téléchargement de toutes les données terminé.")

# --- Étape 6: Création et nettoyage du DataFrame Pandas ---
if all_ohlcv:
    # On crée le DataFrame
    df = pd.DataFrame(all_ohlcv, columns=['Timestamp', 'Open', 'High', 'Low', 'Close', 'Volume'])
    
    # On convertit le timestamp en format de date lisible
    df['Date'] = pd.to_datetime(df['Timestamp'], unit='ms')
    
    # On met la colonne 'Date' comme index du DataFrame
    df.set_index('Date', inplace=True)
    
    # On supprime la colonne 'Timestamp' qui est maintenant redondante
    df.drop('Timestamp', axis=1, inplace=True)

    print("\n--- Aperçu des 5 premières lignes de données ---")
    print(df.head())

    print("\n--- Aperçu des 5 dernières lignes de données ---")
    print(df.tail())

    print("\n--- Informations sur le DataFrame ---")
    df.info()
else:
    print("\nAucune donnée n'a été téléchargée.")
```

```python
import numpy as np

#DataFrame 'df est déjà chargé et contient la colonne 'Close'.

# Étape 1 : Calculer les rendements logarithmiques
df['log_return'] = np.log(df['Close'] / df['Close'].shift(1))

# Étape 2 : Afficher les premières lignes pour vérifier le résultat
print("--- DataFrame avec la nouvelle colonne 'log_return' ---")
print(df.head())

# Étape 3 (Optionnel mais recommandé) : Supprimer la première ligne qui contient une valeur nulle (NaN)
df.dropna(inplace=True)

print("\n--- DataFrame après suppression de la valeur nulle ---")
print(df.head())
```

### Visualiser les séries de prix et de rendements pour observer les régimes de marché.

```python
import matplotlib.pyplot as plt
import seaborn as sns


# --- Configuration du style des graphiques ---
sns.set_style("darkgrid")
plt.rc('figure', figsize=(14, 8))
plt.rc('font', size=12)

# --- Création de la figure et des sous-graphiques ---
# On crée une figure avec 2 graphiques l'un au-dessus de l'autre (2 lignes, 1 colonne)
# sharex=True signifie qu'ils partageront le même axe des x (le temps)
fig, ax = plt.subplots(2, 1, sharex=True)

# --- Graphique 1 : Prix de Clôture ---
ax[0].plot(df['Close'], label='Prix de Clôture', color='skyblue')
ax[0].set_title('Historique des Prix de Clôture du Bitcoin (BTC-USD)')
ax[0].set_ylabel('Prix en USD')
ax[0].legend()

# --- Graphique 2 : Rendements Logarithmiques ---
ax[1].plot(df['log_return'], label='Rendements Logarithmiques', color='salmon')
ax[1].set_title('Rendements Logarithmiques Journaliers du Bitcoin')
ax[1].set_ylabel('Rendement Log')
ax[1].set_xlabel('Date')
ax[1].legend()

# Ajouter une ligne horizontale à zéro pour mieux voir les rendements positifs/négatifs
ax[1].axhline(0, color='grey', linestyle='--')

# --- Affichage ---
plt.tight_layout() # Ajuste automatiquement les graphiques pour qu'ils ne se chevauchent pas
plt.show()
```

## Analyse Exploratoire



```python
print(df['log_return'].describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
import numpy as np

# --- 1. Calculer les paramètres pour la courbe normale ---
# La courbe normale théorique doit avoir la même moyenne et le même écart-type que nos données
mu = df['log_return'].mean()
sigma = df['log_return'].std()

# --- 2. Créer le graphique ---
plt.figure(figsize=(12, 7))

# Tracer l'histogramme des rendements réels
sns.histplot(df['log_return'], 
             bins=100, 
             kde=True,  # Ajoute une estimation lissée de la distribution
             stat="density", 
             color='salmon',
             label='Distribution des Rendements Réels')

# Tracer la courbe de la distribution normale théorique
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000) # Crée une plage de valeurs pour la courbe
pdf = norm.pdf(x, mu, sigma) # Calcule la densité de probabilité pour chaque valeur
plt.plot(x, pdf, 'k--', linewidth=2, label='Distribution Normale Théorique')

# --- 3. Améliorer la lisibilité du graphique ---
plt.title('Distribution des Rendements Logarithmiques du Bitcoin vs. Distribution Normale')
plt.xlabel('Rendement Logarithmique Journalier')
plt.ylabel('Densité')
plt.legend()
plt.show()

# --- 4. Calculer le Kurtosis ---
kurtosis_value = df['log_return'].kurtosis()
print(f"Le kurtosis des rendements est : {kurtosis_value:.2f}")
if kurtosis_value > 3:
    print("Un kurtosis > 3 confirme la présence de 'queues de distribution épaisses' (leptokurtosis).")
```

### Tester la stationnarité des rendements (test ADF).

```python
pip install statsmodels
```

```python
from statsmodels.tsa.stattools import adfuller


# --- 1. Sélectionner la série à tester ---
# Le test ADF s'applique sur une série (une seule colonne)
returns_series = df['log_return']

# --- 2. Réaliser le test ADF ---
print("Réalisation du test Augmented Dickey-Fuller (ADF)...")
# L'option autolag='AIC' permet à la fonction de choisir le nombre de retards optimal
adf_result = adfuller(returns_series, autolag='AIC')

# --- 3. Afficher les résultats de manière claire ---
print(f'ADF Statistic: {adf_result[0]}')
print(f'p-value: {adf_result[1]}')
print('Critical Values:')
for key, value in adf_result[4].items():
    print(f'\t{key}: {value}')

# --- 4. Interpréter le résultat automatiquement ---
print("\n--- Conclusion du test ---")
p_value = adf_result[1]
if p_value <= 0.05:
    print(f"P-value ({p_value:.4f}) <= 0.05. On rejette l'hypothèse nulle.")
    print("Conclusion : La série des rendements est très probablement STATIONNAIRE.")
else:
    print(f"P-value ({p_value:.4f}) > 0.05. On ne peut pas rejeter l'hypothèse nulle.")
    print("Conclusion : La série des rendements est très probablement NON-STATIONNAIRE.")
```

## Modélisation Comparative :

### Model 1 : 30 jours

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Définir les paramètres ---
window_size = 30 # La taille de notre fenêtre glissante (30 jours)

# --- 2. Calculer la volatilité historique sur la fenêtre glissante ---
# .rolling(window=window_size) crée la fenêtre glissante.
# .std() calcule l'écart-type (la volatilité) à l'intérieur de chaque fenêtre.
df['historical_volatility'] = df['log_return'].rolling(window=window_size).std()

# --- 3. Annualiser la volatilité ---
# On multiplie la volatilité journalière par la racine carrée du nombre de jours dans une année.
# Pour les cryptos, qui tradent 365 jours par an, on utilise 365.
annualization_factor = np.sqrt(365)
df['historical_volatility_annualized'] = df['historical_volatility'] * annualization_factor

# --- 4. Afficher les résultats ---
print("--- Aperçu du DataFrame avec la volatilité historique ---")
# Les premières lignes seront NaN car il faut 30 jours de données pour le premier calcul
print(df.tail())

# --- 5. Visualiser la volatilité historique ---
plt.figure(figsize=(14, 7))
plt.plot(df['historical_volatility_annualized'], label='Volatilité Historique Annuelle (30 jours)', color='purple')
plt.title('Volatilité Historique Glissante du Bitcoin')
plt.xlabel('Date')
plt.ylabel('Volatilité Annualisée')
plt.legend()
plt.show()
```

### Modèle 2 : Estimer un modèle GARCH(1,1) sur les rendements.

```python
pip install arch
```

```python
from arch import arch_model
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Préparer les données ---
# On multiplie souvent les rendements par 100 pour aider le solveur du modèle à converger
returns = df['log_return'] * 100

# --- 2. Spécifier et estimer le modèle GARCH(1,1) ---
# vol='Garch' : spécifie un modèle GARCH
# p=1, q=1 : correspond à l'ordre ARCH(1) et GARCH(1)
model = arch_model(returns, vol='Garch', p=1, q=1)

# Estimer le modèle sur les données
# update_freq=5 affiche l'état de l'optimisation toutes les 5 itérations
results = model.fit(update_freq=5)

# --- 3. Afficher le résumé des résultats ---
print("\n--- Résumé du Modèle GARCH(1,1) ---")
print(results.summary())

# --- 4. Visualiser la volatilité prédite par le modèle ---
fig = results.plot(annualize='D')
plt.suptitle('Volatilité Conditionnelle Estimée par GARCH(1,1)')
plt.show()
```

### modèle 3 : Estimer un modèle GJR-GARCH(1,1)

```python
from arch import arch_model
import matplotlib.pyplot as plt

# On réutilise les rendements multipliés par 100.
returns = df['log_return'].dropna() * 100

# --- 1. Spécifier et estimer le modèle GJR-GARCH(1,1) ---
# Le changement est ici : on ajoute o=1 pour inclure le terme asymétrique (gamma)
# p=1 : Terme GARCH (beta)
# o=1 : Terme Asymétrique (gamma)
# q=1 : Terme ARCH (alpha)
model_gjr = arch_model(returns, vol='Garch', p=1, o=1, q=1)

# Estimer le modèle
results_gjr = model_gjr.fit(update_freq=5)

# --- 2. Afficher le résumé des résultats ---
print("\n--- Résumé du Modèle GJR-GARCH(1,1) ---")
print(results_gjr.summary())

# --- 3. Visualiser la volatilité prédite ---
fig = results_gjr.plot(annualize='D')
plt.suptitle('Volatilité Conditionnelle Estimée par GJR-GARCH(1,1)')
plt.show()
```

Cependant, l'interprétation finale des résultats nous apprend que la caractéristique principale de la volatilité du Bitcoin n'est pas l'asymétrie, mais bien l'extrême persistance.

Dans votre modèle GJR-GARCH, **le coefficient beta[1]** est de **0.8625** et il est extrêmement significatif.
